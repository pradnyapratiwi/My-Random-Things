# MENGISI DATA KOSONG

```python
data_hujan[['Sunshine','Humidity3pm', 'Cloud9am', 'Cloud3pm' ]] = data_hujan[['Sunshine','Humidity3pm', 'Cloud9am', 'Cloud3pm' ]].fillna(data_hujan[['Sunshine','Humidity3pm', 'Cloud9am', 'Cloud3pm' ]].mean())
```

```python
data_hujan
```

```python
data_hujan.isnull().sum()
```

```python
data_hujan[['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'WindGustSpeed', 'WindSpeed9am', 
            'WindSpeed3pm', 'Humidity9am','Pressure9am', 'Pressure3pm', 'Temp9am', 
            'Temp3pm']] = data_hujan[['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation',  
                                      'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 
                                      'Pressure9am', 'Pressure3pm', 
                                      'Temp9am', 'Temp3pm']].fillna(data_hujan[['MinTemp', 'MaxTemp', 'Rainfall', 
                                                                                            'Evaporation', 'WindGustSpeed', 
                                                                                            'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am',  
                                                                                            'Pressure9am', 'Pressure3pm','Temp9am', 'Temp3pm']].median())
```
