```python
data_hujan1 = pd.read_csv("hujan.csv") 
data_hujan1
```

```python
MinTemp_mean = data_hujan1['MinTemp'].mean()
data_hujan1['MinTemp'] = data_hujan1['MinTemp'].fillna(MinTemp_mean)
```

```python
MaxTemp_mean = data_hujan1['MaxTemp'].mean()
data_hujan1['MaxTemp'] = data_hujan1['MaxTemp'].fillna(MaxTemp_mean)
```

```python
Rainfall_mean = data_hujan1['Rainfall'].mean()
data_hujan1['Rainfall'] = data_hujan1['Rainfall'].fillna(Rainfall_mean)
```

```python
Evaporation_mean = data_hujan1['Evaporation'].mean()
data_hujan1['Evaporation'] = data_hujan1['Evaporation'].fillna(Evaporation_mean)
```

```python
Sunshine_mean = data_hujan1['Sunshine'].mean()
data_hujan1['Sunshine'] = data_hujan1['Sunshine'].fillna(Sunshine_mean)
```

```python
WindGustSpeed_mean = data_hujan1['WindGustSpeed'].mean()
data_hujan1['WindGustSpeed'] = data_hujan1['WindGustSpeed'].fillna(WindGustSpeed_mean)
```

```python
WindSpeed9am_mean = data_hujan1['WindSpeed9am'].mean()
data_hujan1['WindSpeed9am'] = data_hujan1['WindSpeed9am'].fillna(WindSpeed9am_mean)
```

```python
WindSpeed3pm_mean = data_hujan1['WindSpeed3pm'].mean()
data_hujan1['WindSpeed3pm'] = data_hujan1['WindSpeed3pm'].fillna(WindSpeed3pm_mean)
```

```python
Humidity9am_mean = data_hujan1['Humidity9am'].mean()
data_hujan1['Humidity9am'] = data_hujan1['Humidity9am'].fillna(Humidity9am_mean)
```

```python
Humidity3pm_mean = data_hujan1['Humidity3pm'].mean()
data_hujan1['Humidity3pm'] = data_hujan1['Humidity3pm'].fillna(Humidity3pm_mean)
```

```python
Pressure9am_mean = data_hujan1['Pressure9am'].mean()
data_hujan1['Pressure9am'] = data_hujan1['Pressure9am'].fillna(Pressure9am_mean)
```

```python
Pressure3pm_mean = data_hujan1['Pressure3pm'].mean()
data_hujan1['Pressure3pm'] = data_hujan1['Pressure3pm'].fillna(Pressure3pm_mean)
```

```python
Cloud9am_mean = data_hujan1['Cloud9am'].mean()
data_hujan1['Cloud9am'] = data_hujan1['Cloud9am'].fillna(Cloud9am_mean)
```

```python
Cloud3pm_mean = data_hujan1['Cloud3pm'].mean()
data_hujan1['Cloud3pm'] = data_hujan1['Cloud3pm'].fillna(Cloud3pm_mean)
```

```python
Temp9am_mean = data_hujan1['Temp9am'].mean()
data_hujan1['Temp9am'] = data_hujan1['Temp9am'].fillna(Temp9am_mean)
```

```python
Temp3pm_mean = data_hujan1['Temp3pm'].mean()
data_hujan1['Temp3pm'] = data_hujan1['Temp3pm'].fillna(Temp3pm_mean)
```
